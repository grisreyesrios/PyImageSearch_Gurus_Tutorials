With the following script you can:

1.   Load an image with:
```
cv2.imread()
```
2.   Display the image with: 

```
cv2.imshow()
```

3. Write the image back to the disk and save it with a different format using:

```
cv2.imwrite()
```








````
#Import libraries

import argparse
import cv2 
import sys
sys.path.append("../")
from google.colab.patches import cv2_imshow
````

````
args = {
    "image": "/content/sample_data/cat_silhouette.jpg",
    "output": "/content/sample_data/cat_silhouette_save.jpg"
}
 
#Load image ans show basic information
image_display = cv2.imread(args["image"])
print("width: %d pixels" % (image_display.shape[1]))
print("heigth: %d pixels" % (image_display.shape[0]))
print("channels: %d" % (image_display.shape[2]))
print("Channels are:", image_display.shape)
#Show the image and wait for the keypress
cv2_imshow(image_display)
cv2.imwrite(args["output"], image_display)
````
